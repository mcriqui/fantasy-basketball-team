```python
import csv
from dateutil import parser
from dateutil.tz import *
import pprint
import datetime
```

```python
csv_rows = []

with open('final_schedule.csv', 'r') as fp:
    reader = csv.DictReader(fp)
    for row in reader:
        row.pop('')
        csv_rows.append(row)
```

```python
for row in csv_rows:
    played_on_date = row.get('played_on_date')
    played_on_time = row.get('played_on_time')
#     print(played_on_datetime)

    if played_on_time != "":
        pm_time = "{0} PM".format(played_on_time)
        parsed_time = parser.parse(pm_time)
        date_time_time = parsed_time.time()
        row['played_on_time'] = date_time_time
    else:
        row['played_on_time'] = None
        
    print row['played_on_time']   
    parsed_date = parser.parse(played_on_date)
    date_time_date = parsed_date.date()
    row['played_on_date'] = date_time_date 
    print row['played_on_date']
    
    
```
